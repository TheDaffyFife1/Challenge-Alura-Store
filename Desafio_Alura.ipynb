```python
store1_cat = store1['Categoría del Producto'].value_counts().reset_index()
store1_cat.columns = ['Categoría', 'Ventas']
store1_cat['Tienda'] = 'Tienda 1'

store2_cat = store2['Categoría del Producto'].value_counts().reset_index()
store2_cat.columns = ['Categoría', 'Ventas']
store2_cat['Tienda'] = 'Tienda 2'

store3_cat = store3['Categoría del Producto'].value_counts().reset_index()
store3_cat.columns = ['Categoría', 'Ventas']
store3_cat['Tienda'] = 'Tienda 3'

store4_cat = store4['Categoría del Producto'].value_counts().reset_index()
store4_cat.columns = ['Categoría', 'Ventas']
store4_cat['Tienda'] = 'Tienda 4'

```

```python
df_cat = pd.concat([store1_cat, store2_cat, store3_cat, store4_cat])
```

```python

# Pivotear los datos para el gráfico
pivot_df = df_cat.pivot(index='Categoría', columns='Tienda', values='Ventas')

# Crear el gráfico
plt.figure(figsize=(12, 6))

# Graficar cada tienda
for tienda in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[tienda], marker='o', label=tienda, linewidth=2)

# Personalizar el gráfico
plt.title('Comparación de Ventas por Categoría entre Tiendas', fontsize=14)
plt.xlabel('Categoría del Producto', fontsize=12)
plt.ylabel('Cantidad de Ventas', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Tienda')
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# 4. Productos más y menos vendidos


```python
producto1 = store1['Producto'].value_counts().reset_index()
producto1.columns = ['Producto', 'Ventas']
producto1['Tienda'] = 'Tienda 1'

producto2 = store2['Producto'].value_counts().reset_index()
producto2.columns = ['Producto', 'Ventas']
producto2['Tienda'] = 'Tienda 2'

producto3 = store3['Producto'].value_counts().reset_index()
producto3.columns = ['Producto', 'Ventas']
producto3['Tienda'] = 'Tienda 3'

producto4 = store4['Producto'].value_counts().reset_index()
producto4.columns = ['Producto', 'Ventas']
producto4['Tienda'] = 'Tienda 4'

resultado_final = pd.concat([producto1, producto2, producto3, producto4], ignore_index=True)
```

```python
# Crear un DataFrame vacío para almacenar los resultados completos
df_completo_lista = []

for tienda in tiendas:
    # Obtener los top 10 productos para cada tienda
    top_productos_tienda = resultado_final[resultado_final['Tienda'] == tienda] \
                          .groupby('Producto')['Ventas'].sum() \
                          .nlargest(10).index

    # Crear un DataFrame para esta tienda con sus top productos
    df_tienda = pd.DataFrame({
        'Producto': top_productos_tienda,
        'Tienda': tienda
    })

    df_completo_lista.append(df_tienda)

# Concatenar todos los DataFrames de tiendas
df_completo = pd.concat(df_completo_lista, ignore_index=True)

# Hacer el merge con los datos originales
resultado_completo = pd.merge(df_completo, resultado_final,
                             on=['Producto', 'Tienda'],
                             how='left').fillna(0)

plt.figure(figsize=(14, 10))
colors = plt.cm.tab10.colors  # Paleta de colores

for i, tienda in enumerate(tiendas, 1):
    plt.subplot(2, 2, i)

    # Filtrar y ordenar los productos para esta tienda por ventas (descendente)
    datos_tienda = resultado_completo[resultado_completo['Tienda'] == tienda] \
                 .sort_values('Ventas', ascending=False)

    bars = plt.barh(datos_tienda['Producto'], datos_tienda['Ventas'], color=colors[i-1])
    plt.gca().invert_yaxis()  # Invertir el eje Y para tener el producto con más ventas arriba

    # Añadir etiquetas
    for bar in bars:
        width = bar.get_width()
        if width > 0:  # Solo mostrar etiqueta si hay ventas
            plt.text(width, bar.get_y() + bar.get_height()/2, f'{int(width)}',
                     ha='left', va='center', fontsize=9)

    plt.title(f'Top Productos - {tienda}', fontweight='bold')
    plt.xlabel('Ventas Totales')
    plt.ylabel('Producto')
    plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.suptitle('Top 10 Productos por Tienda', y=1.02, fontsize=14, fontweight='bold')
plt.show()
```
